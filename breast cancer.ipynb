```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import neighbors,preprocessing,svm
import numpy as np
import math
```

```python
df=pd.read_csv("/mnt/c/Users/Srujan/Documents/Datasets/breast-cancer-wisconsin.data")
df.columns=['id','clump_thick','unif_cell_size','unif_cell_shape','marg_adhesion','epithelial_cell_size','bare_nuclei','bland_chromatin','norm_nucleoli','metosis','class']
df.replace('?',-99999,inplace=True)
df.head()
```

```python
df.drop(['id'],1,inplace=True)
df.head()
```

```python
X=np.array(df.drop(['class'],1))
y=np.array(df['class'])
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
preprocessing.scale(X_train)
preprocessing.scale(X_test)
predict_size=math.ceil(0.02*len(X_train))
print(predict_size)
X_lately=X_train[-predict_size:]
X_train=X_train[:-predict_size]
y_train=y_train[:-predict_size]
#clf=neighbors.KNeighborsClassifier()
clf=svm.SVC(C=150,kernel='linear')
#using svm with RBF kernel gives very low accuracy, since there are enough features use linear kernel instead. It gives 
#good accuracy as KNN.
clf.fit(X_train,y_train)
accuracy=clf.score(X_test,y_test)
print(accuracy)
predictions=clf.predict(X_lately)
print(predictions)
```
